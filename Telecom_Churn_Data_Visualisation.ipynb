```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
df = pd.read_csv('https://query.data.world/s/lx5gnlyrog5hozt6sewsw2mqh4jq4s')
df.head()
```

```python
plt.hist(df.MonthlyCharges,bins=20,color='green')
plt.show()
```

```python
sns.distplot(df.MonthlyCharges,hist=False)
plt.show()
```

```python
df1=df[df['Churn']=='Yes']
sns.distplot(df1.MonthlyCharges,hist=False)
plt.show()
```

```python
df2=df[df['Churn']=='No']
sns.distplot(df2.MonthlyCharges,hist=False)
plt.show()
```

```python
sns.distplot(df1.MonthlyCharges,hist=False,label='Churn')
sns.distplot(df2.MonthlyCharges,hist=False,label='NonChurn')
plt.legend()
plt.show()
```

```python
plt.boxplot(df.MonthlyCharges)
plt.show()
```

```python
ten_c=df1['tenure']
ten_nonc=df2['tenure']
s=[['ten_c','ten_nonc']]
plt.boxplot(ten_c)
plt.show()
```

```python
plt.figure(figsize = (10,3))
plt.subplot(1,2,1)
plt.title('Churn')
sns.boxplot(df1['tenure'])

plt.subplot(1,2,2)
plt.title('Non-Churn')
sns.boxplot(df2['tenure'])
```

```python
sns.boxplot(x=df['PaymentMethod'],y=df['TotalCharges'])
plt.xticks(rotation=60)
plt.show()
```

```python
sns.countplot(df['InternetService'],hue=df['Churn'])
plt.xticks(rotation=60)
plt.show()
```
